```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent =  ['the glass of milk',
         'the glass of juice',
         'the cup of tea',
         'I am a good boy',
         'I am a good developer',
         'understand the meaning of words'
         ]
```

```python
#defining the vocab size
vocab_size = 10000
```

```python
one_hot_repr = [one_hot(words,vocab_size) for words in sent]
one_hot_repr
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
sent_length=8
embedded_doccs = pad_sequences(one_hot_repr,padding = 'pre',maxlen=sent_length)
embedded_doccs
```

```python
#feature_representation
dim=10
```

```python
model=Sequential()
model.add(Embedding(10000,dim,input_length=sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embedded_doccs)
```
